## 4. GridSearch

```python
toy_example_domain = pddl.pddl_domain('domain.pddl')
problem_a = pddl.pddl_problem('problem_A.pddl')
problem_b = pddl.pddl_problem('problem_B.pddl')
problem_c = pddl.pddl_problem('problem_C.pddl')
problem_d = pddl.pddl_problem('problem_D.pddl')
problem_e = pddl.pddl_problem('problem_E.pddl')
problem_f = pddl.pddl_problem('problem_F.pddl')
toy_example_problem_list= [problem_a, problem_b, problem_c, problem_d, problem_e, problem_f]
obs_toy_example = pddl.pddl_observations('Observations.csv')
```

```python
model = prap_model.prap_model(toy_example_domain, toy_example_problem_list, obs_toy_example)
```

```python
gs = GridSearch.GridSearch(model, "toy_example_gs")
```

```python
gs.grid
```

```python
gs.add_grid_item([("MOVE_LEFT_FROM", range(5, 10))])
gs.add_grid_item(("MOVE_RIGHT_FROM", range(100, 200)))
gs.add_grid_item(("MOVE_DOWN_FROM", range(200, 3000)))
gs.add_grid_item(("MOVE_LOWER_RIGHT_FROM", range(50, 60)))
gs.add_grid_item(("MOVE_LOWER_LEFT_FROM", range(50, 60)))
gs.create_grid(random=True, size=10)
```

```python
gs.grid
```

```python
gs.check_feasible_domain(multiprocess=True, timeout= 5, 
                         keep_files = False, pickle = True)
```
